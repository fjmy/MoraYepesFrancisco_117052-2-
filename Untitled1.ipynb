18.	Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U y determinar qué fila de U contiene la mayor cantidad de unos. 

```python
import numpy as np #Se importa la libreria numpy, para realizar las 
                   #operaciones necesarias
                  
matriz  =  [[1, 2, 1, 1, 2, 3],
            [3, 1, 1, 1, 5, 2],
            [1, 1, 4, 1, 9, 4],
            [2, 3, 4, 5, 1, 1],
            [3, 1, 2, 3, 1, 1],
            [1, 1, 1, 1, 1, 1]] #Matriz Cuadrada.


print("Matriz cuadrada")

#Ciclo for para imrpimir la matriz
for i in range(len(matriz)):
    print(matriz[i])

#ciclo for para determinar las operaciones en la parte
#de abajo de la matriz
for i in range(len(matriz)):
    #Ciclo for que me idica que la fila deber ser siempre mayor
    #a la columna
    for j in range(len(matriz)):
        if i>j:
            matriz[i][j] = 0


print("Matriz Triangular superior")

#Ciclo for para imprimir la matriz.
for i in range(len(matriz)):
    print(matriz[i])

    
MaxOnes = [] #Lista vacia para contar la cantidad de unos.

#Ciclo for para agregar y contar los unos que posee la matriz. 
for i in range(len(matriz)):
    MaxOnes.append(matriz[i].count(1))

    
print("Las filas con mayor cantidad de unos son:")

#Ciclo for para ubicar la fila donde hay mayor cantidad de unos.
for i in range(len(matriz)):
    if MaxOnes[i] == max(MaxOnes):
        print("Fila :", i+1)
```

19.	Leer una lista e identifique la sublista más larga dentro de dicha lista que se encuentra ordenada ascendentemente. Por ejemplo, en la lista [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la [1, 2, 6, 9]. 

```python
#Lista para realizar el programa
lista = [5, 6, 7, 8, 9, 1, 2, 6, 9, 2, 4, 1]
pos = [] #Lista vacia.


i = 0 #variable para ir guardando la posicion del elemento
      #en la lista.
    
#Ciclo while que me permitira identificar si la lista va en 
#orden ascendente
while i < len(lista)-1:
    if lista[i] < lista[i+1]: #Condicional para identificar 
                              #si va en forma ascendente.
        s = [i, 1]            #sublista, se iran guardando    
                              #los elementos
        k = i  #variable para guardar los elementos.      

        #Ciclo while en donde se guardaran en la sublista los
        #elementos guardados en k
        while k < len(lista)-1:
            if lista[k] < lista[k+1]: #Condicional para agregar el
                                      #elemento k en s.
                s[1] = k+1
            else:
                break
            k += 1
        pos.append(s)
    i += 1

sublist = [] #Sublista vacia
tams = [] #Sublista vacia

for i in range(len(pos)): #Condicion para agregar elementos a la sublista
    sublist.append(lista[pos[i][0]:pos[i][1]+1]) #Comando para agregar los datos
                                                 #a la sublista.
    tams.append(len(sublist[i])) #Comando para agregar los datos a la sublista.

    
#Imprime las listas y sublistas deseadas.    
print("Lista Original:")
print(lista)
print("Sublista ordenada mas larga:")
for i in range(len(sublist)):
    if len(sublist[i]) == max(tams):
        print(sublist[i])
```

20.	Hacer un programa que lea una matriz y la rote 90 grados en sentido antihorario. Imprimir la matriz resultante.
Repita el mismo ejercicio utilizando el módulo numpy. 

```python
matriz  =  [[62, 42, 70, 66, 10],
            [35, 10,  4, 92, 94],
            [36, 37,  5,100, 19],
            [79, 44, 72, 53, 91]]


m = matriz        # m--> matriz de entrada


#Preparo las dimensiones de la matriz de salida mo
mo = [None]*len(m[0])          #nro de columnas de m = nro de columnas mo
for i in range(0,len(m[0])):
    mo[i] = [None]*len(m)      #nro de filas de m = nro de columnas de mo

    


for i in range(len(m)):
    for j in range(-len(m[0]),0):   # empiezo al revés para asegurar el giro antihorario
        mo[-j-1][i] = m[i][j]   # indice de la fila mo ascendente e indice de la columna 
                                #m descendente            

            
## Impresión de las dos matrices
print()
print("Matriz Original")
for i in range(len(m)):
    print(m[i])

print()    
print("Matriz Rotada 90 grados en sentido antihorario")
for i in range(len(mo)):
    print(mo[i])
    
    

## Utilizando numpy.

import numpy as np #Se importa la libreria numpy, para realizar las 
                   #operaciones necesarias

matriz  = np.array([[62, 42, 70, 66, 10],
                    [35, 10,  4, 92, 94],
                    [36, 37,  5,100, 19],
                    [79, 44, 72, 53, 91]]) #Matriz del ejercicio.

M = matriz #Variable donde se guarda la matriz

Mo = (M.T).copy()   # Obtengo la matriz transpuesta

for i in range(Mo.shape[0]):
    Mo[-(i+1),:] = M[:,i].copy()   # Invierto sus columnas

    
print("\nMatriz Original")
print(M)
print("\nMatriz Rotada 90 grados en sentido antihorario")
print(Mo)
```
